```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("tcc_ceds_music.csv.zip")
df
```

```python
 df = df[["artist_name","track_name","release_date","genre","lyrics","topic"]]
df
```

```python
df.rename(columns={"artist_name":"Artist","track_name":"Music_name","release_date":"Release","genre":"Music_Type"},inplace= True)
```

```python
df
```

```python
df["Artist"] = df["Artist"].str.replace(" ","")
```

```python
df ["tags"] = df["lyrics"]+df["Artist"] + df["topic"] + df["Music_Type"]
```

```python
df
```

```python
df = df[["Release","Music_name","tags"]]
df
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
pt = PorterStemmer()
```

```python
def sho(text):
    l = []
    for i in text.split():
        l.append(pt.stem(i))
    return " ".join(l)
```

```python
df["tags"] = df["tags"].apply(sho)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=4500,stop_words="english")
```

```python
vec = cv.fit_transform(df["tags"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cs = cosine_similarity(vec)
```

```python
cs[0]
```

```python
sorted(list(enumerate(cs[0])),reverse=True,key = lambda x:x[1])[1:5]
```

```python
def music_recommend(music):
    music_index = df["Music_name"][df["Music_name"] == music].index[0]
    dis = cs[music_index]
    music_list = sorted(list(enumerate(cs[0])),reverse=True,key = lambda x:x[1])[1:5]

    for i in music_list:
        print(df.iloc[i[0]].Music_name)
```
